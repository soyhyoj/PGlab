## About data 





Mean: flourescent signal recorded from each ROI (Region of Interest)

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches   ## light box
%matplotlib inline
```

```python
data=pd.read_csv("20190314-Exp13.csv", index_col=0, header=0)   #data
print(data.shape)
data.head()
```

```python
# First 30 rows: 1 min record every 2 seconds
# Second 90 rows: every 0.66 seconds
# Last 60 rows: 1 min record every 2 seconds

# Prepare np arrays for each section
first = np.arange(0, 60, 2)
second = np.arange(60, 119, 0.66)
last = np.arange(120, 240, 2)

# Append all the arrays
times = np.append(first, second)
times = np.append(times, last)

# Sanity check: are there 180 time points in total?
len(times)
```

```python
# Column 1 is the actual timestamps of each image taken
# Column 2 is just a sample signal from a random ROI that I created to extract timestamps

data.index = times
data.head()
```

### Normalization of the signals


```python
# Get the average values of the first 30 seconds (F0)
means = data.iloc[:15].mean()
means
```

```python
# Normlize the flourescent signals (normF)
# F0: the average values of the first 30 seconds
# dF = F - F0: difference between the F0 and each point of recording
normF = data.sub(means).div(means)      #dF/F0

#To confirm
#a = data.loc[231,["ROI9"]]
#b =m.loc["ROI9"]
#(a-b)/b

normF.head()
```

```python
normF.plot(legend=False, ylim=[-0.5,2.5], figsize=(12,6), linewidth=1.5)
plt.show()
```

```python
normF.plot(legend=False, ylim=[-0.5,2.5], figsize=(12,6), linewidth=1.5)

plt.xlabel('seconds', fontsize=20)
plt.ylabel('dF/F0', fontsize=20)

## level = y value
## start-end : x (duration) of compound application
## x axis: 1 frame taken every 2 seconds

#level = 1.3
activations = [
    {'start':20,'end':3000, 'color':'grey'},
    ]
#for a in activations:
#    plt.plot([a['start'], a['end']],[level, level], a['color'], linewidth=5)

## label for each compound application
plt.text(5, 2.6, 'DHPG 50uM in bath', color='grey', size=15)

## light
plt.fill([59,59,119,119],[-3.0,12.0,12.0,-3.0], 'salmon')
```

```python
# Individual trace of the first 6 ROIs
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14,14), sharey=True)
cell = ["Mean%d" %x for x in range (1,7)]

for i, ax in enumerate(axes.reshape(-1)):
    ax = normF[cell[i]].plot(ax=ax, ylim=[-1.0, 3.0], color='black', legend=True)
    ax.set_xlabel('seconds', size=15)
    ax.set_ylabel('dF/F0', size=15)
    ax.add_patch(patches.Rectangle((59,-1.5), 60, 10, facecolor='salmon')) # 2PE
```

```python
idxmax = normF[60:150].idxmax()

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14,14), sharey=True)
cell = ["Mean%d" %x for x in range (1,7)]

for i, ax in enumerate(axes.reshape(-1)):
    ax = normF[cell[i]].plot(ax=ax, ylim=[-1.0, 3.0], color='black', legend=True)
    ax.set_xlabel('seconds', size=15)
    ax.set_ylabel('dF/F0', size=15)
    ax.add_patch(patches.Rectangle((59,-1.5), 60, 10, facecolor='salmon')) # 2PE
    ax.vlines(idxmax[i],-2,6, colors='yellow', linewidth=2, linestyles='dotted') # Where lies the max value
```
