```python
import pandas as pd
# Import raw data as DataFrame
columns = ["Cell%d" %x for x in range (1,308)]
filename = "data_405.csv"
data=pd.read_csv(filename, index_col=0, skiprows=1, delimiter=';', header=None, names=columns)   

data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Plot all cell response at once
data.plot(legend=False, ylim=[-0.5,1.5])
```

```python
# Create subplots for each cell response

fig, axes = plt.subplots(nrows=10, ncols=5, figsize=(20, 40), facecolor='white', edgecolor='white')
plt.subplots_adjust(hspace = .3, wspace = .3)
cell = ["Cell%d" %x for x in range (1, 51)]

for i, ax in enumerate(axes.reshape(-1)):
    ax=data[cell[i]].plot(ax=ax, ylim=[-0.5,1.5], color='black', legend=True)
    ax.set_xlabel('time')
    ax.set_ylabel('dF/F0')

    # Annotate where the drug 'X' was applied to the cells
    ax.annotate('X', xy=(118,0.3), xytext=(95, 0.5), size=10, arrowprops=dict(arrowstyle="-|>"))
    ax.annotate('X', xy=(367,1.1), xytext=(343, 1.3), size=10, arrowprops=dict(arrowstyle="-|>"))
    ax.annotate('X', xy=(615,0.3), xytext=(590, 0.5), size=10, arrowprops=dict(arrowstyle="-|>"))
```
